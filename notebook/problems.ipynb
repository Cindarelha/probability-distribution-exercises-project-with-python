# Probability distribution problems

## Binomial distribution

In probability theory and statistics, the binomial distribution is the discrete probability distribution that gives only two possible results in an experiment; for example, when you flip a coin, it can be either head or tails.

```python
from scipy.stats import binom

# Define all the probability functions related to this distribution

def dbinom(x, size, prob = 0.5):
    """
    Calculates the point estimate of the binomial distribution
    """
    result = binom.pmf(k = x, n = size, p = prob, loc = 0)

    return result

def pbinom(q, size, prob = 0.5):
    """
    Calculates the cumulative of the binomial distribution
    """
    result = binom.cdf(k = q, n = size, p = prob, loc = 0)

    return result

def qbinom(p, size, prob = 0.5):
    """
    Calculates the quantile function from the binomial distribution
    """
    result = binom.ppf(q = p, n = size, p = prob, loc = 0)

    return result

def rbinom(n, size, prob = 0.5):
    """
    Generates random variables from the binomial distribution
    """
    result = binom.rvs(n = size, p = prob, size = n)

    return result
```

```python
# Binomial(10, 0.2) distribution, where n = 10 is the number of trials and p = 0.2 is the probability of success in a single trial
import numpy as np

np.random.seed(42)

print(f"Probability that a Binomial(10, 0.2) takes the value 2: {dbinom(2, size = 10, prob = 0.2)}")
print(f"Probability that a Binomial(10, 0.2) will take a value less than 2: {pbinom(2, size = 10, prob = 0.2)}")
print(f"Which value of a Binomial(10, 0.2) has a cumulative probability of 0.9? {qbinom(0.9, size = 10, prob = 0.2)}")
print(f"Generate 2000 random values from a Binomial(10, 0.2) distribution: {rbinom(2000, size = 10, prob = 0.2)}")
```

```python
import matplotlib.pyplot as plt

np.random.seed(42)

size = 10
prob = 0.2
data = rbinom(2000, size = 10, prob = 0.2)

pmf = dbinom(data, size = size, prob = prob)
cdf = pbinom(data, size = size, prob = prob)
ppf = qbinom(data, size = size, prob = prob)

fig, axis = plt.subplots(3, 1, figsize = (5, 7))

axis[0].stem(data, pmf, "r-")
axis[1].stem(data, cdf, "b-")
axis[2].stem(data, ppf, "g-")

# Adjust the layout
plt.tight_layout()

# Show the plot
plt.show()
```

## Exercise 1

Write the above functions but for both normal and uniform distributions.

```python
from scipy.stats import binom, norm,  uniform

# Define all the probability functions related to this distribution

def dnormal(x, mu=0, sigma=1):
    """
    Calculates the point estimate of the norm distribution
    """
    result = norm.pdf(x, loc=mu, scale=sigma)

    return result

def pnormal(q, mu=0, sigma=1):
    """
    Calculates the cumulative of the normal distribution
    """
    result = norm.cdf(q, loc=mu, scale=sigma)
    return result

def qnormal(p, mu =0, sigma = 1):
    """
    Calculates the quantile function from the norm distribution
    """
    result = norm.ppf(p, loc=mu, scale=sigma)

    return result

def rnormal(n, mu =0, sigma = 1):
    """
    Generates random variables from the norm distribution
    """
    result = norm.rvs(loc=mu, scale=sigma, size=n)

    return result

import matplotlib.pyplot as plt

# Seed for reproducibility
np.random.seed(42)

# Parameters
mu = 0
sigma = 1

# Generate data
data = rnormal(2000, mu, sigma)

# Calculate pmf (pdf in case of normal distribution), cdf, and quantiles
pmf = dnormal(data, mu, sigma)
cdf = pnormal(data, mu, sigma)
ppf = qnormal(data, mu, sigma)

# Create subplots
fig, axis = plt.subplots(3, 1, figsize=(10, 15))

# Plot PMF (PDF in case of normal distribution)
axis[0].stem(data, pmf, 'r-')


# Plot CDF
axis[1].stem(data, cdf, 'b-')


# Plot PPF
axis[2].stem(data, ppf, 'g-')


# Adjust the layout
plt.tight_layout()

# Show the plot
plt.show()
```

```python
from scipy.stats import uniform

def duniform(x, loc,  scale):
    """
    Calculates the probability density function of the uniform distribution
    """
    result = uniform.pdf(x, loc=loc, scale=scale)
    return result

def puniform(q, loc, scale):
    """
    Calculates the cumulative distribution function of the uniform distribution
    """
    result = uniform.cdf(q, loc=loc, scale=scale)
    return result

def quniform(p, loc, scale):
    """
    Calculates the quantile function (percent-point function) of the uniform distribution
    """
    result = uniform.ppf(p, loc=loc, scale=scale)
    return result

def runiform(n, loc, scale):
    """
    Generates random variables from the uniform distribution
    """
    result = uniform.rvs(loc=loc, scale=scale, size=n)
    return result

import numpy as np
import matplotlib.pyplot as plt

# Seed for reproducibility
np.random.seed(42)

# Parameters
loc = -2
scale = 1

# Generate data
data = runiform(50, loc, scale)

# Calculate pdf, cdf, and quantiles
pdf = duniform(data, loc, scale)
cdf = puniform(data, loc, scale)
ppf = quniform(data, loc, scale)

# Create subplots
fig, axis = plt.subplots(3, 1, figsize=(10, 15))

# Plot PDF
axis[0].stem(data, pdf, 'r-')


# Plot CDF
axis[1].stem(data, np.sort(cdf), 'b-')


# Plot PPF
axis[2].stem(data, ppf, 'g-')


# Adjust the layout
plt.tight_layout()

# Show the plot
plt.show()
```
